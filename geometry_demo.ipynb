Importiere Module und definiere Einstellungen f√ºr die Plots

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math

def show_plot():
    # set plot settings
    ax = plt.gca()
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.set_xlim([-3, 11])
    ax.set_ylim([-3, 11])
    ax.grid(True)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
    plt.show()

```

Definiere und Plotte die nicht-rotierten Punkte

```python
initial_point = (1,1)

n_points_per_line = 10
primary_points = [(initial_point[0] + i , initial_point[1]) for i in range (n_points_per_line)]
print("primary_line: " + str(primary_points))

secondary_points = [(x,-y) for x,y in primary_points]
print("secondary_line: " + str(secondary_points))

def plot_points(points):
    points_x = [x for x,y in points]
    points_y = [y for x,y in points]
    plt.plot(points_x, points_y, "o")

plot_points(primary_points)
plot_points(secondary_points)
show_plot()
```

Rotiere alle Punkte und Plotte erneut

```python

def rotate_point(point, angle_degrees):
    x, y = point
    angle_radians = math.radians(angle_degrees)
    x_rotated = x * math.cos(angle_radians) - y * math.sin(angle_radians)
    y_rotated = x * math.sin(angle_radians) + y * math.cos(angle_radians)
    return (x_rotated, y_rotated)

rotation_degree = 10
rotated_primary_points = [rotate_point(p, rotation_degree) for p in primary_points]
rotated_secondary_points = [rotate_point(p, rotation_degree) for p in secondary_points]

plot_points(rotated_primary_points)
print("rotated_primary_points: " + str(rotated_primary_points))
plot_points(rotated_secondary_points)
print("rotated_secondary_points: " + str(rotated_secondary_points))

show_plot()
```
